# a :

#### فرض کنید می خواهیم آرایه زیر را مرتب کنیم

```python
a = np.array([[4 , 5 , 7 , 1],
              [7 , 9 , 3 , 5],
              [2 , 7 , 1 , 6],
              [8 , 4 , 0 , 5]])
a
```

# kind :
### ( 'quicksort' , 'heapsort' , 'mergesort' , 'stable' )
### نوشتن این پارامتر اختیاری است و نوع الگوریتمی که برای مرتب سازی استفاده می شود را مشخص می کند و یکی از الگوریتم های بالا را می پذیرد که در

## 'quicksort' :
این الگوریتم برای مرتب سازی سریع و بهینه آرایه های بزرگ و چند بعدی به کار می رود و در بیشتر مواقع به ویژه در خصوص آرایه های بزرگ از سایر الگوریتمها سریعتر عمل می کند روش کار این الگوریتم این است که به صورت متناوب یک آرایه را به آرایه های کوچکتر تقسیم کرده و پس از مرتب کردن آنها در نهایت آنها را با هم ترکیب میکند تا مجدداً
تبدیل به یک آرایه شوند 
اگر بخواهیم عملکرد این الگوریتم را دقیقتر توضیح دهیم به این صورت است که از میان عناصر موجود در یک آرایه یکی از آنها را که بهتر است میانه مقادیر باشد به عنوان نقطه پیوت
انتخاب کرده و بقیه مقادیر را به دو گروه مقادیر کمتر و بیشتر از آن نقطه تقسیم کرده و برای هر گروه یک ماتریس جدید تشکیل داده و در دو طرف ماتریسی که 
تنها عنصر آن مقدار پیوت انتخاب شده است ، قرار میدهد. سپس همین عملیات را بار دیگر بر روی هر یک از ماتریسهای ساخته شده به صورت متناوب آنقدر تکرار میکند تا
تمام مقادیر مرتب شوند و در نهایت ماتریسهای ساخته شده را با هم ترکیب میکند و یک ماتریس واحد با مقادیر مرتب شده میسازد

# 


```python
np.sort(a , kind = 'quicksort' ) 
```

# 'heapsort' :
این الگوریتم نوعی الگوریتم مبنتنی بر مقایسه است که ابتدا با ساختن یک درخت هیپ دودویی از آرایه مورد نظر شروع میکند. به این صورت که در لایه اول یک مقدار نوشته شده و در لایه بعد دو عدد از سمت چپ به راست به صورت زیرشاخه لایه قبلی قرار میگیرد. لایه بالاتر  والد و لایه های پایین تر فرزندان هستند.ابتدا یک مقدار از سمت چپ زیرشاخه والد قرار میگیرد و با آن مقایسه میشود. اگر از آن بزرگتر بود جای دو مقدار با هم عوض میشود اما اگر کوچکتر بود به همان صورت باقی مانده و مقدار بعدی سمت راست لایه اضافه می شود. سپس مقادیر بعدی مجدداً به لایه های پایین تر از چپ به راست اضافه شده و زیر شاخه لایه دوم قرار میگیرند و هر مقدار با سرشاخه و والد خود مقایسه میشود و اگرمقدار لایه های پایینی بزرگتر بود با لایه مقدار والد جا به جا میشود. شاخه ها آنقدر لایه به لایه اضافه و مقایسه میشوند که در نهایت شکل کلی درخت شبیه یک هرم خواهد بود. این کار آنقدر ادامه پیدا میکند تا تمام مقادیر مرتب شوند و سپس خروجی به صورت یک ماتریس مرتب شده نمایش داده میشود.  

```python
np.sort(a , kind = 'heapsort' ) 
```

# 'mergesort':
این الگوریتم نیز مانند کوییک سورت از استراتژی تقسیم و حل استفاده میکند. در این استراتژی یک مسئله بزرگ به ابعاد کوچکتر و قابل حل تقسیم می شود و هر بخش به صورت جداگانه حل شده و در نهایت نتایج با هم ترکیب می شوند. در این الگوریتم نیز، ابتدا لیست اولیه به دو بخش تقسیم می‌شود، سپس به صورت بازگشتی و متناوب برای هر یک از این بخش‌ها مرحله تقسیم را تکرار می کند تا هر بخش فقط شامل یک عنصر باشد.در مرحله بعد، عملیات مرتب‌سازی آغاز می‌شود. در این مرحله، بخش‌های تقسیم شده با هم مقایسه شده و به صورت مرتب شده به هم پیوند می‌شوند. برای این کار، از دو اشاره‌گر استفاده می‌شود که به ابتدای هر بخش اشاره می‌کنند. سپس اشاره‌گرها به صورت متوالی برای مقایسه دو عنصر به کار می‌روند و کوچکترین عنصر به لیست جدید اضافه می‌شود. اشاره‌گر به سمت راست حرکت می‌کند و مرحله مقایسه و اضافه کردن به لیست جدید تکرار می‌شود تا تمامی عناصر به لیست جدید اضافه شوند. در آخرین مرحله، لیست جدید به عنوان لیست مرتب‌شده بازگردانده می‌شود. 

```python
np.sort(a , kind = 'mergesort' ) 
```

# 'stable' :
در الگوریتم‌های مرتب‌سازی، ممکن است برخی عناصر با هم برابر باشند. در زمان انجام عملیات مرتب سازی ممکن است موجب تغییر ترتیب آنها شویم. در این شرایط، اگر مرتب‌سازی انجام شده، به همان شکل که بوده است باقی بماند و ترتیب عناصری که با هم برابر هستند حفظ شود، الگوریتم مرتب‌سازی به عنوان "استیبل" شناخته می‌شود. به طور مثال، اگر در لیستی از شخصیت‌ها دو شخص به نام "آلیس" وجود داشته باشد ، در حین مرتب سازی، اگر الگوریتم مرتب سازی استیبل باشد، ترتیب دو شخص "آلیس" حفظ خواهد شد و عنصری که اولین بار وارد لیست شده است، قبل از عنصری که بعداً وارد لیست شده است، قرار خواهد گرفت. این بدان معناست که ترتیب جدید، نسبت به ترتیب قبلی، استیبل خواهد بود. در مقابل، اگر الگوریتم مرتب‌سازی استیبل نباشد، ممکن است ترتیب عناصری که با هم برابر هستند تغییر کند. به عنوان مثال، در الگوریتم مرتب‌سازی تعداد شماره‌های موبایل، اگر دو شماره موبایل با پیش‌شماره متفاوت ولی با شماره برابر وجود داشته باشد، اگر الگوریتم مرتب‌سازی استیبل باشد، ترتیب آنها باقی می‌ماند ولی در غیر این صورت، ترتیب آنها ممکن است تغییر کند.

```python
np.sort(a , kind = 'stable' ) 
```

# order : 
ورود این پارامتر نیز اختیاری است. اگر یک آرایه دارای فیلدهای معین باشد، با استفاده از این پارامتر می توانیم مشخص کنیم که مرتب سازی اول بر اساس کدوم فیلد ، دوم بر اساس کدام فیلد و ... باید انجام شود
 :به طور مثال فرض کنید آرایه ای به صورت زیر داریم که ستون اول آن نام، ستون دوم سن و ستون سوم قد افراد است

```python
arr = np.array([('John' , 28, 175),
                ('Mary' , 32, 163),
                ('David', 24, 180),
                ('Sarah', 29, 170)] , dtype=[('name', 'S5'), ('height', int), ('age', int)] )
print(arr)
```

######
اکنون فرض کنید قصد داریم آرایه بالا را بر اساس قد و سپس بر اساس سن مرتب کنیم

```python
sorted_arr = np.sort(arr, order=['height','age'])
print(sorted_arr)
```
